```python
from sklearn.datasets import load_iris
```

```python
import numpy as np
import pandas as pd
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
dt = load_iris()
```

```python
dt.keys()
```

```python
X,Y = dt.data,dt.target
```

```python
dt.target_names
```

```python
#i/p layer 4 and hidden layer 30 # hidden layer 30 and 10 #hidden layer 10 and o/p layer 3
model = nn.Sequential (
    nn.Linear(4, 30),
    nn.ReLU(),
    nn.Linear(30, 10),
    nn.ReLU(),
    nn.Linear(10, 3),
    #nn.Softmax(dim = 1)
)
```

```python
X = torch.FloatTensor(X)
Y = torch.LongTensor(Y)
```

```python
#model(X)
```

```python
#torch.argmax(model(X) , dim=1)
```

```python
lossfn = nn.CrossEntropyLoss()
optSgd = optim.SGD(model.parameters(), lr =0.01)
```

Training


```python
for _ in range(200):
    optSgd.zero_grad()
    Yp = model(X)
    loss = lossfn(Yp, Y)
    loss.backward()
    optSgd.step()

    print(loss.item())
```

Inference

```python
ynewp = model(X)
```

```python
sftout = torch.softmax(ynewp,dim=1)
```

```python
df = pd.DataFrame(X)
```

```python
df['Y'] = Y
```

```python
data = torch.tensor(df.to_numpy())
```

```python
model1 = nn.Sequential (
    nn.Linear(4, 30),
    nn.ReLU(),
    nn.Linear(30, 10),
    nn.ReLU(),
    nn.Linear(10, 3),
    #nn.Softmax(dim = 1)
)
```

```python
lossfn = nn.CrossEntropyLoss()
batchSgd1 = optim.SGD(model1.parameters(), lr =0.01)
```

```python
batches = DataLoader(data, batch_size=10,shuffle = True)
```

```python
model1.train()
for epoch in range(200):
  batch_loss = 0
  for i, batch in enumerate(batches):
    batch_X = batch[:, :-1].float()
    batch_Y = batch[:, -1].long()

    batchSgd1.zero_grad()
    batch_Yp = model1(batch_X)
    loss = lossfn(batch_Yp, batch_Y)
    loss.backward()
    batchSgd1.step()
    batch_loss = batch_loss + loss.item()
    print(f"Epoch {epoch+1}:  Total Loss = {batch_loss}")

```
